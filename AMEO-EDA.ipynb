```python
lst = []
for i in df['10board']:
    if isinstance(i, str):
        if 'icse' in i.lower():
            lst.append('ICSE')
        elif 'cbse' in i.lower() or 'central board of secondary education' in i.lower() or 'delhi' in i.lower() or 'cbsc' in i.lower():
            lst.append('CBSE')
        else:
            lst.append('State Board')
    else:
        lst.append('State Board')
```

```python
df['10board'] = lst
df['10board'].value_counts()
```

```python
pd.pivot_table(data = df,values='Salary', index = '10board').plot(kind='bar') #students graduated from ICSE board are earning high salaries.
```

```python
pd.pivot_table(data = df,values='Salary', index = '12board').plot(kind='bar')
```

```python
list1 = []
for i in df['12board']:
    if isinstance(i, str):
        if 'isc' in i.lower():
            list1.append('ISC')
        elif 'cbse' in i.lower():
            list1.append('CBSE')
        elif 'nios' in i.lower():
            list1.append('National Board')
        elif 'diploma' in i.lower():
            list1.append('Diploma')
        elif 'technical' in i.lower():
            list1.append('Technical')
        elif 'university' in i.lower():
            list1.append('University')
        else:
            list1.append('State Board')
    else:
        list1.append('State Board')
```

```python
df['12board']=list1
df['12board'].value_counts()
```

```python
pd.pivot_table(data = df,values='Salary', index = '12board').plot(kind='bar')# more salary is earned by the ISC board 
```
